```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
sns.set() # make matplotlib far less ugly
```

## Sample datasets

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

## Line plots

```python
fig, ax =plt.subplots(1,2, figsize=(13, 8.27))
figsize=(15,4)
sns.lineplot(x="sepal_length", y="sepal_width", hue="species", 
             data=iris, palette="bright", ax=ax[0]) \
    .set_title('Iris')

sns.lineplot(x="culmen_depth_mm", y="culmen_length_mm", hue="species", 
             data=penguins, palette="bright", ax=ax[1]) \
    .set_title('Penguins')
```
